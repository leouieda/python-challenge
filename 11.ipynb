```python
%matplotlib inline
import requests
import re
from IPython.display import Image
from skimage import io
import matplotlib.pyplot as plt
```

```python
url = 'http://www.pythonchallenge.com/pc/return/{}'
# Use the username and password from challenge 8
auth = ('huge', 'file')
r = requests.get(url.format('5808.html'), auth=auth)
page = r.text
print(page)
```

Download the image.

```python
imgfile = 'cave.jpg'
imgpath = 'files/{}'.format(imgfile)
r = requests.get(url.format(imgfile), auth=auth)
print("Status code:", r.status_code)
with open(imgpath, 'wb') as f:
    for chunk in r:
        f.write(chunk)
```

```python
Image(imgpath)
```

Separate the image into even and odd pixels.

```python
img = io.imread(imgpath)
even = img[::2, ::2]
odd = img[1::2, 1::2]

plt.figure()
plt.imshow(even)
plt.title('Even')
plt.figure()
plt.imshow(odd)
plt.title('Odd')
```
