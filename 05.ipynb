# Challenge: http://www.pythonchallenge.com/pc/def/peak.html

```python
from urllib import request
import pickle
```

Get the HTML for this challenge.

```python
url = 'http://www.pythonchallenge.com/pc/def/{}'
```

```python
with request.urlopen(url.format('peak.html')) as response:
    # decode transforms the bytes read into unicode
    page = response.read().decode()
print(page)
```

```python
fname, headers = request.urlretrieve(url=url.format('banner.p'), filename='files/banner.p')
print(fname)
```

```python
with open(fname, mode='rb') as f:
    banner = pickle.load(f)
print(banner)
```

```python
for line in banner:
    for char, n in line:
        print(char*n, end='')
    print()
```
