```python
%matplotlib inline
import requests
import re
from IPython.display import Image
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'http://www.pythonchallenge.com/pc/return/{}'
# Use the username and password from the previous challenge
auth = ('huge', 'file')
r = requests.get(url.format('good.html'), auth=auth)
page = r.text
print(page)
```

```python
first = np.array([int(i) for i in ''.join(page.splitlines()[25:-9]).split(',')])
print('Size:', len(first))
print(first)
```

```python
second = np.array([int(i) for i in ''.join(page.splitlines()[-7:-2]).split(',')])
print('Size:', len(second))
print(second)
```

Get the image from the website.

```python
r = requests.get(url.format('good.jpg'), auth=auth)
print("Status code:", r.status_code)
with open('files/good.jpg', 'wb') as f:
    for chunk in r:
        f.write(chunk)
```

Turns out the image has nothing to do with the challenge. Plot the coordinate pairs for the first and second part.

```python
x1, y1 = first.reshape((first.size//2, 2)).T
x2, y2 = second.reshape((second.size//2, 2)).T

plt.figure()
plt.plot(x1, y1, '-m', lw=2)
plt.plot(x2, y2, '-c', lw=2)
plt.ylim(400, 100)
```
